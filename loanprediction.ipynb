```python
df.loc[df.Dependents=='3+','Dependents']=3
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)
df['Dependents'] = df['Dependents'].astype('int64')
df['Credit_History'].fillna(df['Credit_History'].mode()[0],inplace=True) 
df['Credit_History'] = df['Credit_History'].astype('int64')
```

```python
  def count_func(df,col):
    categories = df[col].dropna().unique()
    data = {}
    for cat in categories:
        data[cat] = len(df[df[col]==cat])
    data = pd.DataFrame(data,index=[0]).T
    data.columns = [col]
    print(data)
    return data                                                                                                                                         
```

```python
count_df = count_func(df,'Gender')
count_df.plot(kind='bar',title='Gender Count') #using pandas
plt.show()
```

```python
plt.bar(list(count_df.index), list(count_df[count_df.columns[0]])) #using matplotlib
plt.title('Gender count') # adding title
plt.show()
```

```python
cnt_married_stat = count_func(df,'Married')
plt.bar(list(cnt_married_stat.index), list(cnt_married_stat[cnt_married_stat.columns[0]])) #using matplotlib
plt.xlabel("Married Status") #adding xlabel
plt.ylabel('Frequency/Counts') #adding ylabel
plt.show()
```

```python
cnt_ed_stat = count_func(df,'Education')
plt.pie(list(cnt_ed_stat[cnt_ed_stat.columns[0]]), labels = list(cnt_ed_stat.index))
plt.title('Comparing Education Qualification')
plt.show()
```

```python
cnt_depend_stat = count_func(df,'Dependents')
plt.pie(list(cnt_depend_stat[cnt_depend_stat.columns[0]]), labels = list(cnt_depend_stat.index), explode = [0,0,0.1,0],
       autopct='%1.1f%%') #explode portion of pie and adding percentage composition
plt.title('Comparing No Dependants of loan applicant')
plt.show()
```
